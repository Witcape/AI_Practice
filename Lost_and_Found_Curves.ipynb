```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
models = ['YOLOv10 + SAM',
    'Grounding DINO + SAM',
    'Mask2Former',
    'OneFormer (SWIN-L)',
    'OneFormer (DINAT-L)']

clip_32 = {
    'Accuracy': [32.14, 45.45, 54.55, 36.36, 36.36],
    'Precision': [14.87, 20.78, 18.18, 10.39, 15.58],
    'Recall': [11.64, 20.51, 17.95, 10.26, 15.38]
}

clip_16 = {
    'Accuracy': [38.79, 63.64, 45.45, 36.36, 36.36],
    'Precision': [5.91, 18.18, 7.79, 10.39, 7.79],
    'Recall': [7.68, 17.95, 7.69, 10.26, 7.69]
}

# Create DataFrame
df = pd.DataFrame({
    'Model': models * 2,
    'CLIP': ['CLIP-32'] * len(models) + ['CLIP-16'] * len(models),
    'Accuracy': clip_32['Accuracy'] + clip_16['Accuracy'],
    'Precision': clip_32['Precision'] + clip_16['Precision'],
    'Recall': clip_32['Recall'] + clip_16['Recall']
})

# Plotting each metric separately with updated style settings
metrics = ['Accuracy', 'Precision', 'Recall']
x = np.arange(len(models))
width = 0.35

for metric in metrics:
    fig, ax = plt.subplots(figsize=(10, 6))
    values_32 = df[df['CLIP'] == 'CLIP-32'][metric].values
    values_16 = df[df['CLIP'] == 'CLIP-16'][metric].values

    ax.bar(x - width/2, values_32, width, label='CLIP-32')
    ax.bar(x + width/2, values_16, width, label='CLIP-16')

    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, fontsize=14)
    ax.set_ylabel('Score (%)', fontsize=16)
    ax.set_title(f'{metric}: CLIP-32 vs CLIP-16 for Textual Query', fontsize=16)
    ax.legend(title='CLIP Version', fontsize=10, title_fontsize=10)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xlabel('Model', fontsize=16)
    ax.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
from pinecone import Pinecone
import numpy as np
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set global font and style
mpl.rcParams["font.family"] = "DejaVu Sans"
mpl.rcParams["axes.edgecolor"] = "white"
mpl.rcParams["axes.facecolor"] = "white"
mpl.rcParams["figure.facecolor"] = "white"

# 1) Initialize Pinecone client
pc = Pinecone(api_key="pcsk_2fjuni_CrEgFxrDhXfZmyXGq58zKzWoFq5FWAYXuaQhtpuHW5ZBGqjZPANzzriV4w94Rg8")

# 2) Choose your index
index_name = "dino16"
index = pc.Index(index_name)

# 3) Get total vector count
stats = index.describe_index_stats()
total_vectors = stats["total_vector_count"]
print(f"Total vectors in '{index_name}': {total_vectors}")

# 4) Truncate if too many
max_k = 10000
top_k = min(total_vectors, max_k)
if total_vectors > max_k:
    print(f"⚠️  total_vectors ({total_vectors}) > {max_k}, truncating to top_k = {max_k}")

# 5) Pull IDs using zero-vector
dummy_vector = np.zeros(512).tolist()
resp = index.query(
    vector=dummy_vector,
    top_k=top_k,
    include_values=False,
    include_metadata=False
)
ids = [m["id"] for m in resp["matches"]]
print(f"Fetched {len(ids)} IDs")

# 6) Fetch embeddings in batches
batch_size = 100
emb_list = []
for i in range(0, len(ids), batch_size):
    batch_ids = ids[i : i + batch_size]
    fetch_resp = index.fetch(ids=batch_ids)
    for vec in fetch_resp.vectors.values():
        emb_list.append(vec.values)

embeddings = np.array(emb_list)
print("Embeddings shape:", embeddings.shape)

# 7) K-means clustering
n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)

# 8) t-SNE projection
tsne = TSNE(n_components=2, random_state=42)
proj = tsne.fit_transform(embeddings)

# 9) Sexy scatter plot
fig, ax = plt.subplots(figsize=(12, 10))

colors = plt.cm.tab10.colors  # nice soft palette of 10

for cluster_id in np.unique(labels):
    mask = (labels == cluster_id)
    ax.scatter(
        proj[mask, 0],
        proj[mask, 1],
        s=30,
        alpha=0.8,
        color=colors[cluster_id % len(colors)],
        label=str(cluster_id + 1)
    )

# Remove axes, spines, ticks
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

# Equal aspect ratio
ax.set_aspect('equal', 'box')

# Beautiful legend: 2 rows, centered, no border
legend = ax.legend(
    ncol=5,
    bbox_to_anchor=(0.5, -0.08),
    loc='upper center',
    frameon=False,
    fontsize=12,
    title="Cluster",
    title_fontsize=13
)

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.show()
```
