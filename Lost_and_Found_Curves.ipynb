```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
models = ['YOLOv10 + SAM',
    'Grounding DINO + SAM',
    'Mask2Former',
    'OneFormer (SWIN-L)',
    'OneFormer (DINAT-L)']

clip_32 = {
    'Accuracy': [32.14, 45.45, 54.55, 36.36, 36.36],
    'Precision': [14.87, 20.78, 18.18, 10.39, 15.58],
    'Recall': [11.64, 20.51, 17.95, 10.26, 15.38]
}

clip_16 = {
    'Accuracy': [38.79, 63.64, 45.45, 36.36, 36.36],
    'Precision': [5.91, 18.18, 7.79, 10.39, 7.79],
    'Recall': [7.68, 17.95, 7.69, 10.26, 7.69]
}

# Create DataFrame
df = pd.DataFrame({
    'Model': models * 2,
    'CLIP': ['CLIP-32'] * len(models) + ['CLIP-16'] * len(models),
    'Accuracy': clip_32['Accuracy'] + clip_16['Accuracy'],
    'Precision': clip_32['Precision'] + clip_16['Precision'],
    'Recall': clip_32['Recall'] + clip_16['Recall']
})

# Plotting each metric separately with updated style settings
metrics = ['Accuracy', 'Precision', 'Recall']
x = np.arange(len(models))
width = 0.35

for metric in metrics:
    fig, ax = plt.subplots(figsize=(10, 6))
    values_32 = df[df['CLIP'] == 'CLIP-32'][metric].values
    values_16 = df[df['CLIP'] == 'CLIP-16'][metric].values

    ax.bar(x - width/2, values_32, width, label='CLIP-32')
    ax.bar(x + width/2, values_16, width, label='CLIP-16')

    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, fontsize=14)
    ax.set_ylabel('Score (%)', fontsize=16)
    ax.set_title(f'{metric}: CLIP-32 vs CLIP-16', fontsize=16)
    ax.legend(title='CLIP Version', fontsize=10, title_fontsize=10)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xlabel('Model', fontsize=16)
    ax.grid(True)

    plt.tight_layout()
    plt.show()
```
